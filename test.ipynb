```python
from tavily import TavilyClient

# 1. Initialize the client with your API Key
tavily = TavilyClient(api_key="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")

def perform_tavily_search(query):
    # 2. Executing the search
    # search_depth="advanced" includes more detailed content if needed
    response = tavily.search(query=query, search_depth="basic", max_results=3)
    
    # 3. Process and display results
    print(f"Results for: {query}\n" + "="*40)
    
    for result in response['results']:
        title = result.get('title')
        url = result.get('url')
        content = result.get('content')
        
        print(f"Title: {title}")
        print(f"URL: {url}")
        print(f"Snippet: {content[:150]}...") # Truncated for readability
        print("-" * 20)

# --- Execute ---
user_query = "What are the top AI news stories today?"
perform_tavily_search(user_query)
```
